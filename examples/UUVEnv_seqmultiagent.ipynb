# UUV Multi-Agent Environment visiting Targets Sequentially

## Creating the environment

```python
# initialize setup
from fimdpenv import setup, UUVEnv
setup()
```

```python
# create envionment
from fimdpenv.UUVEnv import SynchronousMultiAgentEnv
env = SynchronousMultiAgentEnv(num_agents=3, grid_size=[20,20], capacities=[50,50,50], reloads=[22,297], targets=[67, 345, 268, 153], init_states=[350, 178, 65])
env
```

```python
# get consmdp
mdp, targets = env.get_consmdp()
```

```python
# allocate targets
allocation_list = [[67, 153], [268], [345]]
env.allocate_targets(allocation_list)
env.targets_alloc
```

```python
# list reload states
env.reloads
```

## Simulating and visualizing trajectories

```python
env.animate_seqsimulation(rth=False, interval=100, save=False)
```

```python
env.target_histories
```

## Selecting agent colors and Return-to-Home

```python
import numpy as np
colors = [np.array([1,0,0]), np.array([0,1,0]), np.array([0,0,1])]

env = SynchronousMultiAgentEnv(num_agents=3, grid_size=[20,20], capacities=[50,50,50], reloads=[22,297], targets=[67, 345, 268], init_states=[350, 178, 65], agents_colors=colors)

# allocate targets - one agent has no allocation
allocation_list = [[67],[],[268, 345]]
env.allocate_targets(allocation_list)

# animate
env.animate_seqsimulation(rth=True, interval=50, save=False)
```

```python
env.target_histories
```

```python
env.state_histories[1]
```

## Enhanced action space

```python
# create envionment with enchanced action space
from fimdpenv.UUVEnv import SynchronousMultiAgentEnv
env2 = SynchronousMultiAgentEnv(num_agents=3, grid_size=[20,20], capacities=[50,50,50], reloads=[22,297], targets=[67, 345, 268], init_states=[350, 178, 65], enhanced_actionspace=1)

# allocate targets
allocation_list = [[67], [268], [345]]
env2.allocate_targets(allocation_list)

env2
```

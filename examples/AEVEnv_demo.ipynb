# AEV Environment Demo in the streets of NYC

```python
from fimdpenv import AEVEnv
import fimdp
from fimdp.energy_solver import GoalLeaningES
```

```python
# create environment
env = AEVEnv.AEVEnv(capacity=100, target=['42440465'], init_state='42434894')
```

```python
# create consmdp and fenerate strategy
mdp, targets = env.get_consmdp()
solver = GoalLeaningES(mdp, env.capacity, targets, threshold=0.1)
strategy = solver.get_strategy(fimdp.energy_solver.BUCHI)
env.update_strategy(strategy)
```

```python
# run simulation and visualize a strategy
env.update_environment(strategy)
env.visualize_strategy()
```

Blue represents reload states, green represents initial state, red represents target state and black line is the path.

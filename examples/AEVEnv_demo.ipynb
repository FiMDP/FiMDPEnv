# AEV Environment Demo in the streets of NYC

```python
from fimdpenv import AEVEnv
import fimdp
```

We first define reload states and target states. If not specified explicitly, the relaod states are assumed to be all electric charging stations in the area of interest. Once we specify the reload and target states, we also need to specify the agent's capacity and its initial state while creating an instance of AEV environment. 

```python
# definre reloads and targets
reloads = ['42431659','42430367','1061531810','42443056','1061531448','42448735','596775930','42435275','42429690','42446036','42442528','42440966','42431186','42438503','42442977',
 '42440966','1061531802','42455666']
targets = ['42440465','42445916']
```

```python
# create environment
env = AEVEnv.AEVEnv(capacity=1000, targets=targets, reloads=reloads, init_state='42434894')
env.reset()
```

Once we create an instance of the environment, we create a counterstrategy using the create_counterstrategy() method. This method creates the counter strategy and stores it as the strategy attribute of the instance.

```python
# create strategy
env.create_counterstrategy()
```

You can view a snapshot of the environment at any time instant using the render_map() method. The following render shows the initial status of the environment. 

```python
env.render_map()
```

Blue points represents reload states, green point represents initial state, and red points represents target state. The black line shows the trajectory of the agent obtained by following the energy-aware counter-selector.

One could also run an instance of the simulation using the generated strategy and plot the corresponding trajectory. This can be done using the animate_simulation() method demonstrated below. The method accepts two inputs namely the number of simulation time steps and the interval between each frame in the animation. 

```python
env.animate_simulation(num_steps=1000, interval=10)
```
